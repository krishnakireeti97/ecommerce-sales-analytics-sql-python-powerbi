```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

%matplotlib inline

DATA_DIR = Path("../outputs")
CLEAN_FILE = DATA_DIR / "cleaned_sales_data.csv"

CLEAN_FILE
```

```python
df = pd.read_csv(CLEAN_FILE)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
total_orders = df["order_id"].nunique()
total_customers = df["customer_id"].nunique()
total_revenue = df["net_amount"].sum()

total_orders, total_customers, total_revenue
```

```python
city_sales = (
    df.groupby("city", as_index=False)["net_amount"]
      .sum()
      .sort_values("net_amount", ascending=False)
)

city_sales
```

```python
category_sales = (
    df.groupby("category", as_index=False)["net_amount"]
      .sum()
      .sort_values("net_amount", ascending=False)
)

category_sales
```

```python
payment_sales = (
    df.groupby("payment_method", as_index=False)["net_amount"]
      .sum()
      .sort_values("net_amount", ascending=False)
)

payment_sales
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["order_month_year"] = df["order_date"].dt.to_period("M").astype(str)

monthly_sales = (
    df.groupby("order_month_year", as_index=False)["net_amount"]
      .sum()
      .sort_values("order_month_year")
)

monthly_sales
```
